```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = sns.load_dataset('diamonds')
df = pd.DataFrame(data)
```

```python
df.head()
```

## Pearson Correlation Coefficient (r-value)
- Measures the type and strength of the linear correlation between two variables.
- The number ranges between 1.0 and -1.0.
- When the r-value is close to 1.0, it indicates a strong positive correlation.
- When the r-value is close to -1.0, it indicates a strong negative correlation.
- When the r-value is 0.00, it indicates no correlation.

```python
num_data = data.select_dtypes(include=['float64', 'int64'])
num_data.corr().head(7)
```

```python
sns.heatmap(data=num_data.corr()[['price']].sort_values(by='price', ascending=False),
            annot=True, cmap='Greens', cbar=False, fmt=f'.3f')
```

## Procedure for Creating and Using a Regression Model
1. Split dataset.
2. Create model from training dataset.
3. Validate model with test dataset.
4. If model is valid, predict data with it.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    num_data[['carat']], num_data[['price']],
    test_size=0.30, random_state=20)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
y_predicted = model.predict(x_test)
y_predicted
```

```python
predicted = pd.DataFrame(y_predicted, columns = ['Predicted_Price'])
predicted.head(3)
```

```python
final = predicted.join([x_test.reset_index(drop=True),
                        y_test.reset_index(drop=True)])
final.head()
```

```python
finalMelt = pd.melt(final, id_vars=['carat'], 
                    value_vars=['price','Predicted_Price'], 
                    var_name='Price_Type', value_name='Price_Value')
finalMelt
```

```python
sns.relplot(data=finalMelt, kind='line', x='carat', y='Price_Value', hue='Price_Type')
```

## Regression line equation is **y = mx + b**, where...
- y is the **dependent** variable
- m is the **slope** of the line
- x is the **independent** variable
- b is the **Y-intercept**

```python
X = data.carat.values.reshape(-1,1)
y = data.price
```

```python
lr = LinearRegression()
predicted = lr.fit(X,y)
```

```python
m = predicted.coef_
b = predicted.intercept_
print(f'The slope is {m} and the y-intercept is {b}')
```

```python
print(f'Regression line equation is y = {m} * x + {b}')
```

```python
#Estimate the price using the regression equation
carat_val = 0.70
estimate_price = (m * carat_val) + b
estimate_price_scalar = estimate_price.item()
print(f'The predicted price for a {carat_val:.2f} carat diamond is ${estimate_price_scalar:,.2f}')
```

```python
carat_val = 0.54
estimate_price = (m * carat_val) + b
estimate_price_scalar = estimate_price.item()
print(f'The predicted price for a {carat_val:.2f} carat diamond is ${estimate_price_scalar:,.2f}')
```
